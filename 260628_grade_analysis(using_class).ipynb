```python
datas =[]
```

```python
#데이터를 int로 변환 후 2차원 리스트에 저장.

with open(file= file_path, mode='r') as f:
  for idx, line in enumerate(f):

    if idx >0:
      y= list(map(int,line.split(',')))
      # print(idx, y)
      datas.append(y)
```

```python
class Subject:
  def __init__(self, datas, id=0, class_id =0, sub =0):
    self.datas = datas
    self.subject = sub

    #아이디정보
    self.id =id
    self.class_id = class_id

    #과목점수 리스트
    self.math =[]
    self.english =[]
    self.science =[]


  #1. 과목 번호를 입력하면 리스트를 만들어서 리턴
  #   classdata = 클래스별 성적리스트를 만들어논 리스트.
  def sub_lists(self, subject, classdata =[]):
    # 클래스 기준으로 그룹화한 성적 리스트가 없다면, 과목별 리스트 출력.
    if len(classdata) ==0:
       classdata =self.datas

    #실수로 여러번 실행해서 리스트가 늘어나지 않도록 리셋
    self.math =[]
    self.english =[]
    self.science =[]


    # 과목을 숫자로 구분. 1 = 수학, 2 = 영어, 3 = 과학
    match subject:
      case 1:
        for x in classdata:
          self.math.append(x[2])
        print('과목: [수학]')
        return self.math
      case 2:
        for x in classdata:
          self.english.append(x[3])
        print('과목: [영어]')
        return self.english
      case 3:
        for x in classdata:
          self.science.append(x[4])
        print('과목: [과학]')
        return self.science
      case _:
        return []

  #2. 과목 번호를 입력하면 해당 과목의 최댓값, 최솟값, 평균  리턴
  def total_subject(self, subject=0, classlist=[]):
     # 클래스 기준으로 그룹화한 성적 리스트가 없다면, 과목별로 계산.
    if len(classlist) == 0:
       checkdata =self.sub_lists(subject)
    else:
      checkdata = classlist

    max, min, total = 0,checkdata[0],0

    # print(checkdata)
    for x in checkdata:
      total += x

      if x> max:
        max =x
      if x <min:
        min =x

    mean =total/len(checkdata)

    return f'최댓값: {max}, 최솟값: {min}, 평균: {mean}'

  #3. 반과 과목 번호를 입력하면 최댓값, 최솟값, 평균, 총점 리턴
  def total_class(self, class_id, subject):
    max, min, total = 0,0,0

    checkdata=[]

    #1. 선택한 반으로 데이터 담기. 없거나 1~5반 사이가 아니면 담지않음.
    for x in self.datas:
      if class_id == 0:
        print('반을 입력하지 않으셨습니다.')
        return []
      elif type(class_id) != int:
        print('반은 1~5반까지 있습니다.')
        return []
      elif x[1] == class_id:
        checkdata.append(x[:])

    print(f'[{class_id}반]')
    #print(checkdata)

    #2. 기존의 과목별 리스트를 만드는 함수 이용.
    total_list = self.sub_lists(subject,checkdata)

    #3. 최댓값, 최솟값, 평균 만드는 함수 이용.
    total_val = self.total_subject(classlist=total_list)

    #4. 총점만 따로 계산
    for x in total_list:
      total += x

    return f'총점: {total} {total_val}'
```

```python
# 1(수학), 2 (영어), 3 (과학)을 입력하면 리스트를 출력.

# 수학점수의 리스트를 출력
test1 =Subject(datas)
print(test1.sub_lists(1))


# 영어 점수의 리스트를 출력
print(test1.sub_lists(2))

# 과학점수의 리스트를 출력
print(test1.sub_lists(3))
```

```python
# # 1(수학), 2 (영어), 3 (과학)을 입력하면 최댓값, 최솟값, 평균을 출력.
# 수학점수의 최댓값, 최솟값, 평균
print(test1.total_subject(1))

# 영어점수의 최댓값, 최솟값, 평균
print(test1.total_subject(2))

# 과학점수의 최댓값, 최솟값, 평균
print(test1.total_subject(3))
```

```python
# 1반 수학점수의 값
# Subject(datas, sub =1,class_id=1)
print(test1.total_class(1,1))


# 1반 영어점수의 값
# Subject(datas, sub =2,class_id=1)
print(test1.total_class(1,2))



# # 1반 과학점수의 값
# Subject(datas, sub =2,class_id=1)
print(test1.total_class(1,3))
```
