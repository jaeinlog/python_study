#02. try-except

```python

# try:
#   (1) 실행할 코드 블록
# except [에러이름 [as 별명]]:           # 에러이름 생략 시 모든 종류에러 처리가능
#   (2) 예외 상황이 발생했을 때 예외를 처리하기 위한 코드 블록. 여러개 작성 가능
#           # 에러이름 지정 시 선언된 에러들만 처리할 수 있음.
# [
# else:
#   (3) try 블록의 모든 코드들이 예외(오류)없이 정상 수행이 됐을 때 실행할 코드블록
# finally:
#   (4) 예외(오류) 발생 여부와 상관없이 항상 실행할 코드 블록
# ]

# * 대괄호[]부분은 제외처리 가능.

#try 구문의 실행 순서
# 정상실행: (1) --> (3) --> (4)
# 예외(오류) 발생: (1) --> (2)
```

# 2-1. try-except 예시
숫자를 입력받았을 때 오류 발생을 한다는 상황을 가정하여, 예외처리 방법을 작성.

```python
# 해당 자판기는 1~5를 입력하면 해당 음료수를 출력하게되어있다.
# 이외의 값을 입력하면 다시 입력하라는 문구를 띄워주는 코드를 작성할 것.

drinks = {1:'밀크커피', 2:'원두커피', 3:'콜라',4:'사이다',5:'반환'}

#01. 에러명을 작성하여 각각 처리.
# except 별명을 모두 e로 처리해도 그 블락에서만 처리되서 별명이 모두 동일해도 상관없음
# 참고: 별명을 쓰는 이유는 print(e)하면 오류 확인이 쉬워서 사용
try:
    print('------------------------[자판기]------------------------')
    choice_drink = int(input('[숫자]로 음료수를 골라주세요! >>>'))
    print(f'[{drinks[choice_drink]}] 를 고르셨습니다. ')
except ValueError as e:                           #int에 문자열 실수 입력 시 ValueError
  print('문자/실수를 입력하셨습니다. 다시입력하세요!')
except KeyError as e:                             #dict에 없는 키 값 입력 시 KeyError
  print('해당 메뉴에 없는 번호를 선택하셨습니다. 다시입력하세요!')
else:
  if choice_drink!=5:
    print('출력완료. 맛있게 드세요~!')
  else:
    print('환불되었습니다.')
finally:
  print('종료')
```

# 02-1. 에러명을 작성하여 모두 처리.

```python
try:
    print('------------------------[자판기]------------------------')
    print(drinks)
    choice_drink = int(input('[숫자]로 음료수를 골라주세요! >>>'))
    print(f'[{drinks[choice_drink]}] 를 고르셨습니다. ')
 except Exception as e:                                  # Exception : 모든 오류
    print('입력오류! 다시입력하세요!')
else:
  if choice_drink!=5:
    print('출력완료. 맛있게 드세요~!')
  else:
    print('환불되었습니다.')
finally:
  print('종료')
```

#02-2. 에러명을 작성하지않고 모두처리.

```python
try:
    print('------------------------[자판기]------------------------')
    print(drinks)
    choice_drink = int(input('[숫자]로 음료수를 골라주세요! >>>'))
    print(f'[{drinks[choice_drink]}] 를 고르셨습니다. ')
 except:                                           # 별명으로 출력안할거면 이렇게표시해도됨.
    print('입력오류! 다시입력하세요!')
else:
  if choice_drink!=5:
    print('출력완료. 맛있게 드세요~!')
  else:
    print('환불되었습니다.')
finally:
  print('종료')
```

#03-1. while _무한루프하여 정상적인 값 입력 시 멈추기.

```python
drinks = {1:'밀크커피', 2:'원두커피', 3:'콜라',4:'사이다',5:'반환'}
# 위 자판기를 1~5의 값을 입력해야만 멈추는 코드로 변경했다.
while True:
  try:
      print('------------------------[자판기]------------------------')
      print(drinks)
      choice_drink = int(input('[숫자]로 음료수를 골라주세요! >>>'))
      print(f'[{drinks[choice_drink]}] 를 고르셨습니다. ')
  except ValueError as e:
    print('문자/실수를 입력하셨습니다. 다시입력하세요!')
  except KeyError as e:
    print('해당 메뉴에 없는 번호를 선택하셨습니다. 다시입력하세요!')

  else:
    if choice_drink!=5:
      print('출력완료. 맛있게 드세요~!')
    else:
      print('환불되었습니다.')
    break                     #코드가 예외처리없이 정상적으로 처리되면 멈추기
```

#03-2. for _무한루프하여 정상적인 값 입력 시 멈추기.

```python
for _ in iter(int, 1):
  try:
      print('------------------------[자판기]------------------------')
      print(drinks)
      choice_drink = int(input('[숫자]로 음료수를 골라주세요! >>>'))
      print(f'[{drinks[choice_drink]}] 를 고르셨습니다. ')
  except:
      print('입력오류! 다시입력하세요!')
  else:
    if choice_drink!=5:
      print('출력완료. 맛있게 드세요~!')
    else:
      print('환불되었습니다.')
    break                   #코드가 예외처리없이 정상적으로 처리되면 멈추기
```
